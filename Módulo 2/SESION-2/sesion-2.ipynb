### Crear un archivo CSV

```python
with open('data/my-data.csv', 'w', newline = "") as file:
    writer = csv.writer(file)

    writer.writerow(['ID', 'Pais', 'Ciudad', 'Continente'])
    writer.writerow(['1', 'Francia', 'Paris', 'Europa'])
    writer.writerow(['2', 'Belgica', 'Bruselas', 'Europa'])
```

## PokeApi

```python
url = 'https://pokeapi.co/api/v2/pokemon/4'
response = requests.get(url)
```

```python
pokemon = response.json()
```

```python
pokemon['name']
```

```python
pokemon['sprites']
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
plt.imshow(mpimg.imread(pokemon['sprites']['front_default']))
plt.title(pokemon['name'])
```

```python
plt.figure(figsize=(10,10))
for i in range(20):
    plt.subplot(5, 4, i+1)

    url = 'https://pokeapi.co/api/v2/pokemon/{}'.format(i+1)
    response = requests.get(url)
    pokemon = response.json()
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(mpimg.imread(pokemon['sprites']['front_default']))
    plt.title(pokemon['name'])

plt.show()
```
